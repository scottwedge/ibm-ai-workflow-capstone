# Business Scenario

_"Our customers will be able to subscribe and pay for AAVAIL services by combining any services they desire and being charged only for those services."_

**Manager statement:** <br>
_"AAVAIL product managers will be able to project a particular countryâ€™s revenue for the following month using a machine learning model that will provide revenue estimates and confidence measures for those estimates."_

**Business opportunity:** <br>
We can have a ML model that works at least as well as the current one (in terms of accuracy), saving the manager's time.

Past sales data is correlated with future sales data <br>
A ML model can work as well as the current one (in terms of accuracy)

Revenue data, split by country and month: <br>

| Date | Country | Revenue |
| --- | --- | --- |

# Data Analysis

**Goals** <br>
1 - Investigate the relationship between the relevant data, the target and the business metric <br>
2 - Articulate findings using a deliverable with visualizations

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('train-data/train.csv')
df.head()
```

Do we need customer_id, day, invoice? - We'll assume 'no'. <br>
Is the total_price = price * times_viewed ? - We'll assume 'yes'.<br>

```python
df['revenue'] = df['price'] * df['times_viewed']
df['date'] = pd.to_datetime((df.year*10000+df.month*100+df.day).apply(str),format='%Y%m%d')
```

```python
print('There are ' + str(len(np.unique(df.country))) + ' countries in the dataset.')
```

We will take the top 10 countries by revenue, and group the rest as "Others"

```python
top_countries = [country for country in df.groupby(['country']).agg({'revenue':'sum'})\
                 .sort_values('revenue', ascending = False).index[:10]]

df['country_group'] = df.country.apply(lambda x: x if x in top_countries else 'others')
```

```python
# we will get time series dataframes for each country
agg_dict = {'customer_id' : 'nunique',
            'invoice' : 'nunique',
            'price' : 'mean',
            'revenue':'sum',
            'stream_id' : 'nunique',
            'times_viewed':'sum'
           }
country_dict = {}

for country in np.unique(df['country_group']):
    df_temp = df[df['country_group']==country].resample('D', on='date').agg(agg_dict)
    df_temp = df_temp.fillna(0)
    df_temp['country'] = country
    
    country_dict[country] = df_temp
```

```python
for key in country_dict.keys():
    df_temp = country_dict[key]
    df_temp['revenue_past_3days'] = df_temp.rolling('3d')['revenue'].sum()
    df_temp['revenue_past_7days'] = df_temp.rolling('7d')['revenue'].sum()
    df_temp['revenue_past_30days'] = df_temp.rolling('30d')['revenue'].sum()
    df_temp['target'] = df_temp.rolling('30d')['revenue'].sum().shift(-30) #revenue in the next 30 days
    country_dict[key] = df_temp
```

```python
df = pd.concat([df for df in country_dict.values()], axis=0)
```

```python
# plot data
fig, ax = plt.subplots(figsize=(15,7))

df.groupby([df.index.year, df.index.month, df.country]).revenue.sum().unstack().plot(ax=ax)
plt.show()
```

We can't see much because of United Kingdom. Let's look again, without it:

```python
fig, ax = plt.subplots(figsize=(15,7))
df.groupby([df.index.year, df.index.month, df.country]).revenue.sum().unstack().drop('United Kingdom', axis = 1).plot(ax=ax)
plt.show()
```

Looking for correlations

```python
df.corr()
```

```python
df.plot.scatter('revenue_past_3days', 'target')
df.plot.scatter('revenue_past_7days', 'target')
df.plot.scatter('revenue_past_30days', 'target')
```

# Modelling

We don't necessarily need to approach this problem as a time series one. Instead, we can take past revenue as a feature and use regression models. <br>

Finally, we can see there is a lot of heteroskedasticity, so we can start by a robust linear model.

**Pipeline:** <br>
1 - Train/test split <br>
2 - Encoding <br>
3 - Standard scaling <br>
4 - Model training and hyperparameter tuning <br>
5 - Picking a model <br>

**Models to try:** <br>
1 - Robust linear regression <br>
2 - CART <br>
3 - Random Forest <br>
4 - XGBoost  <br>

```python
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer, make_column_transformer

from statsmodels.regression.linear_model import OLS
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb

from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error as mse
```

## Train/test split

```python
df = df[~df['target'].isna()]

X = df.drop(['target','country'], axis = 1)
y = df['target']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.8, random_state=42)
```

## Pre-processing

```python
numeric_cols = list(X_train)

preprocess = make_column_transformer(
    (StandardScaler(), numeric_cols)
)

X_train = preprocess.fit_transform(X_train)
X_test = preprocess.fit_transform(X_test)
```

## Modelling

### Robust linear regression

```python
ols1 = OLS(y_train,X_train).fit(cov_type='HC1')
ols1_result = ols1.predict(X_test)
print(ols1.summary())
```

### CART

```python
# Grid search
parameters = {'splitter':('random', 'best'),
              'max_depth':[None, 10, 100],
              'max_features':['auto','sqrt','log2',0.3],
              'presort':[True]
             }

model = DecisionTreeRegressor()
clf = GridSearchCV(model, parameters, cv = 5, iid = False)
clf.fit(X_train, y_train)

cart = clf.best_estimator_
cart_result = cart.predict(X_test)
```

### Random Forest

```python
# Grid search
parameters = {'n_estimators':[3, 5, 10, 20, 50, 100],
              'criterion':['mae', 'mse'],
              'max_depth':[None, 10, 100],
              'max_features':['auto','sqrt','log2',0.3]
             }

model = RandomForestRegressor()
clf = GridSearchCV(model, parameters, cv = 5, iid = False)
clf.fit(X_train, y_train)

rforest = clf.best_estimator_
rforest_result = rforest.predict(X_test)
```

### XGBoost

```python
model = xgb.XGBRegressor()

parameters = {'objective':['reg:squarederror'],
              'learning_rate': [0.05, 0.1],
              'max_depth': [1, 5, 50],
              'subsample': [0.7, 0.9],
              'colsample_bytree': [0.7, 0.9],
              'n_estimators': [100, 1000],
              'seed': [42]}

clf = GridSearchCV(model, parameters, cv = 5, iid = False)

clf.fit(X_train, y_train)

xgboost = clf.best_estimator_
xgboost_result = xgboost.predict(X_test)
```

```python
result_table = pd.DataFrame({'y':y_test,
                             'ols1':ols1_result,
                             'cart':cart_result,
                             'rforest':rforest_result,
                             'xgboost':xgboost_result})

result_table
```

```python
print('MSE')
for col in result_table.columns:
    print(col + ': ' + str(round(mse(result_table['y'], result_table[col]), 0)))
```

```python
print('Best model:')
print(min({col: mse(result_table['y'], result_table[col]) for col in result_table.columns.drop('y')}.items(),
          key=lambda x: x[1]))
```
